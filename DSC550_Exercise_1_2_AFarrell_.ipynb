
## 1. Load the dataset as a Pandas data frame

```python
# importing libraries needed
import numpy as np
import pandas as pd
# loading dataset as a data frame
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
```

## 3. Find the dimensions
Finding the number of rows and columns in the data frame.

## 4. Find the top five games by critic score.

```python
# Sort the DataFrame by "Critic_Score" in descending order and get the top 5 rows
top_5 = df.sort_values(by="Critic_Score", ascending=False).head(5)

# Display only the "Name" and "Critic_Score" columns
top_5_short = top_5[["Name", "Critic_Score"]]


print("Top 5 games by critic score:")
print(top_5_short)
```

## 5. Find the number of video games in the data frame in each genre.

```python
# Count the number of video games in each genre
genre_count = df["Genre"].value_counts()

print("Number of video games in each genre:")
print(genre_count)
```

## 6. Find the first five games in the data frame on the SNES platform.

```python
# Ensure Release_Date is in datetime format
df["Year_of_Release"] = pd.to_datetime(df["Year_of_Release"])

# Filter for SNES games and sort by Release_Date
snes_games_sorted = df[df["Platform"] == "SNES"].sort_values(by="Year_of_Release").head(5)

# Display only the game Title
first_snes_games = snes_games_sorted[["Name"]]

print("First five SNES games based on year of release:")
print(first_snes_games)
```

## 7. Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.

```python
# First grouping by publisher and calculating the total global sales for each
total_publisher_sales = df.groupby("Publisher")["Global_Sales"].sum()

# Sort the publishers by total global sales in descending order and get the top 5
top_publishers = total_publisher_sales.sort_values(ascending=False).head(5)

print("Top 5 publishers by total global sales:")
print(top_publishers)
```

## 8. Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.

```python
# Calculating the percentage of global sales from North America (NA_Sales)
df["NA_Percentage"] = (df["NA_Sales"] / df["Global_Sales"]) * 100

# Display the first five rows of the new DataFrame
print("DataFrame with NA Percentage column:")
print(df.head(5))
```

## 9. Find the number NaN entries (missing data values) in each column.

## 10. Calculate the median user score of all the video games

Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
# Identifying all non-numerical string values in  the User_Score by attempting to convert to numeric
non_numerical_count = df["User_Score"].apply(pd.to_numeric, errors="coerce").isna().sum()

print(f"Count of non-numerical string values in User_Score column: {non_numerical_count}")
```

```python
# Since this count is higher than the numnber of NaN values, we know that we need to replace the non-numerical values in order to calculate the medium.

# Finding and replacing all non-numerical string values within the user score column with 'NaN' since they cannot be converted to a float or used in the calculation.
# We can do this by converting 'user_Score' to numeric then replacing all non-convertible strings with 'NaN'.
df["User_Score"] = pd.to_numeric(df["User_Score"], errors="coerce")

# Testing replacment on a game we know had a non-numerical value ('tbd').
Zumba_Fitness = df[df["Name"] == "Zumba Fitness"]
Zumba_Fitness2 = Zumba_Fitness[["Name", "User_Score"]]
print(Zumba_Fitness2)
```

```python
# Remove rows any with NaN in the User_Score column so we can calculate the median.
# filtered_NaN_df = df.dropna(subset=["User_Score"])
```

```python
# Calculating the median User_Score for each game.
grouped_df = df.groupby("Name")["User_Score"].median()
grouped_df
```

```python
# Map the median scores back to the original DataFrame using Title as the key.
df["Median_Score"] = df["Name"].map(grouped_df)

# Replace NaN values in User_Score with the median score for each game.
df["User_Score"] = df["User_Score"].fillna(df["Median_Score"])

# Dropping the Median_Score "helper column" from the orginal data frame.
df = df.drop(columns=["Median_Score"])
```

```python
# Re-counting all non-numerical string values in the User_Score after filling in the median for each game.
non_numerical_count = df["User_Score"].apply(pd.to_numeric, errors="coerce").isna().sum()

print(f"Count of non-numerical string values in User_Score column: {non_numerical_count}")
```

```python
# Testing replacment on a game we know had a non-numerical value ('tbd').
Zumba_Fitness = df[df["Name"] == "Zumba Fitness"]
Zumba_Fitness2 = Zumba_Fitness[["Name", "User_Score"]]
print(Zumba_Fitness2)
```
